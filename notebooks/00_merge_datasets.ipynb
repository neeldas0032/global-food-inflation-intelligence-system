```python
import pandas as pd
```

```python
df2020 = pd.read_csv("../data/wfp_food_prices_global_2020.csv")
df2021 = pd.read_csv("../data/wfp_food_prices_global_2021.csv")
df2022 = pd.read_csv("../data/wfp_food_prices_global_2022.csv")
df2023 = pd.read_csv("../data/wfp_food_prices_global_2023.csv")
df2024 = pd.read_csv("../data/wfp_food_prices_global_2024.csv")
df2025 = pd.read_csv("../data/wfp_food_prices_global_2025.csv")
df2026 = pd.read_csv("../data/wfp_food_prices_global_2026.csv")
```

```python
merged_df = pd.concat([
    df2020,
    df2021,
    df2022,
    df2023,
    df2024,
    df2025,
    df2026
], ignore_index=True)
```
